```python
import random
```

```python
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

puppy = '''
              _=,_
           o_/6 /#\\
           \__ |##/
            ='|--\\
              /   #'-.
              \#|_   _'-. /
               |/ \_( # |" 
              C/ ,--___/

'''
```

```python
your_num = int(input("What do you choose? Type 0 for Rock, 1 for Paper, or 2 for Scissors.\n"))

com_num = random.randint(0, 2)

choice_map = {0: rock, 1: paper, 2: scissors}

if your_num not in [0, 1, 2]:
    print(puppy)
    print("A puppy jumps out and begs you to play the game the right way.")
else:
    print('You choose:')
    print(choice_map.get(your_num))
    print('Computer chooses:')
    print(choice_map.get(com_num))
    if (your_num - com_num == 1) or (your_num == 0 and com_num == 2):
        print('Congratulation! You win!')
    elif your_num == com_num:
        print('It is a draw.')
    elif (com_num - your_num == 1) or (com_num == 0 and your_num == 2):
        print('You lose.')
```
