```python
import colorgram
from turtle import Turtle, Screen
from random import choice
```

```python
# Extract 10 colors from an image.
colors = colorgram.extract('hirst_image.jpg', 30)

color_list = []

for color in colors:
    color_list.append((color.rgb[0], color.rgb[1], color.rgb[2]))
```

```python
color_list = color_list[4:]
color_list[:3]
```

```python
def set_position(timmy):
    timmy.setpos(-250, -250)

def next_line(timmy):
    timmy.penup()
    timmy.left(90)
    timmy.forward(50)
    timmy.left(90)
    timmy.forward(500)
    timmy.right(180)
```

```python
timmy = Turtle()
timmy.speed("fastest")
timmy.penup()
timmy.hideturtle()

set_position(timmy)


my_screen = Screen()
my_screen.colormode(255)

for _ in range(10):
    for _ in range(10):
        timmy.dot(20, choice(color_list))
        timmy.forward(50)
    next_line(timmy)

my_screen.exitonclick()
```
