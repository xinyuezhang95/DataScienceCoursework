```python
import math 
from IPython.display import clear_output

logo = """
 ______________________
|  __________________  |
| | XinyueZhang   0. | |  .----------------.  .----------------.  .----------------.  .----------------. 
| |__________________| | | .--------------. || .--------------. || .--------------. || .--------------. |
|  ___ ___ ___    ___  | | |     ______   | || |      __      | || |   _____      | || |     ______   | |
| | 7 | 8 | 9 |  | + | | | |   .' ___  |  | || |     /  \     | || |  |_   _|     | || |   .' ___  |  | |
| |___|___|___|  |___| | | |  / .'   \_|  | || |    / /\ \    | || |    | |       | || |  / .'   \_|  | |
| | 4 | 5 | 6 |  | - | | | |  | |         | || |   / ____ \   | || |    | |   _   | || |  | |         | |
| |___|___|___|  |___| | | |  \ `.___.'\  | || | _/ /    \ \_ | || |   _| |__/ |  | || |  \ `.___.'\  | |
| | 1 | 2 | 3 |  | x | | | |   `._____.'  | || ||____|  |____|| || |  |________|  | || |   `._____.'  | |
| |___|___|___|  |___| | | |              | || |              | || |              | || |              | |
| | . | 0 | = |  | / | | | '--------------' || '--------------' || '--------------' || '--------------' |
| |___|___|___|  |___| |  '----------------'  '----------------'  '----------------'  '----------------' 
|______________________|
"""
```

```python
def add(n1, n2):
    return n1 + n2

def subtract(n1, n2):
    return n1 - n2

def multiply(n1, n2):
    return n1 * n2

def divide(n1, n2):
    return n1 / n2

def square(n1):
    return n1 ** 2

def sqrt(n1):
    return math.sqrt(n1)

operations = {
    "+": add,
    "-": subtract,
    "*": multiply,
    "/": divide,
    "square": square,
    "square root": sqrt,
}
```

```python
def calculator():
    print(logo)

    num1 = float(input("What's the first number?: "))
    for symbol in operations:
        print(symbol)
    should_continue = True

    while should_continue:
        operation_symbol = input("Pick an operation: ")
        if operation_symbol in ["square", "square root"]:
            calculation_function = operations[operation_symbol]
            answer = calculation_function(num1)
            print(f"The {calculation_function.__name__} of {num1} is {answer}")
        else:
            if operation_symbol in operations.keys():
                num2 = float(input("What's the next number?: "))
                calculation_function = operations[operation_symbol]
                answer = calculation_function(num1, num2)
                print(f"{num1} {operation_symbol} {num2} = {answer}")
            else:
                print("You typed an invalid operation symbol. Calculation ended.")
                should_continue = False
                break
        
        if_continue = input(f"Type 'y' to continue calculating with {answer}, or type 's' to start a new calculation, or type 'n' to end the calculator: ")

        if  if_continue == 'y':
            num1 = answer
        elif if_continue == 's':
            should_continue = False
            clear_output()
            calculator()
        elif if_continue == 'n':
            print("Calculation ended.")
            should_continue = False
        else:
            print("You typed an invalid input. Calculation ended.")
            should_continue = False
```

```python
calculator()
```
